```python
import docsGpt
from pathlib import Path
import os

INPUT_PATH = Path.cwd() / 'input'
OUTPUT_PATH = Path.cwd() / 'output'
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)

POSTDOC_PATH = INPUT_PATH / 'PostDoc'

PHD_PATH = INPUT_PATH / 'PhD'
PHD_OUTPUT_FILEPATH = OUTPUT_PATH / 'results_phd.csv'
PHD_OUTPUT_ERROR_FILEPATH = OUTPUT_PATH / 'errors_phd.csv'

with open('prompt_phd.txt', 'r') as f:
    PROMPT_PHD = f.read()
    PROMPT_PHD_N_COLUMNS = len(PROMPT_PHD.split('|'))
```

```python
# count = 0
# max_count = 3
separator = '|'

# Loop over all files in subfolders
for root, dirs, files in os.walk(PHD_PATH):
    for filename in files:
        # Extract name and country of candidate from path
        root = Path(root)
        name = root.stem
        country = root.parent.stem
        # Process file if is a PDF (it contains the CV info)
        if filename.endswith('.pdf'):
            # Loop until the result contain PROMPT_PHD_N_COLUMNS elements
            n_columns = 0
            while n_columns != PROMPT_PHD_N_COLUMNS:
                # Prepare the query
                filepath = root / filename
                query = f"Extracts information, in CSV format separated by '{separator}', based on this {len(PROMPT_PHD.split(';'))}-element header: 'INFO'. Fill in any missing data with '-'. The name is 'NAME' and the current country of residence is 'COUNTRY'. Fill in the gender with 'Male', 'Female' or 'Unknown'. The result output must be {len(PROMPT_PHD.split(';'))} values separated by '{separator}', not including the initial header."
                query = query.replace('INFO', PROMPT_PHD)
                query = query.replace('NAME', name)
                query = query.replace('COUNTRY', country)

                # Send the query to ChatGPT and get the result response
                try:
                    wrapped_text = docsGpt.extract_info_from_cv(
                        str(filepath), query)
                except Exception as e:
                    print(e)
                    with open(PHD_OUTPUT_ERROR_FILEPATH, 'a') as fe:
                        fe.write(f'{str(filepath)}: {e}\n')

                # Process the result response
                result = ' '.join(wrapped_text).replace(
                    f' {separator}', separator).replace(f'{separator} ', separator).strip()

                # Update n_columns from result response
                n_columns = len(result.split(separator))
                print(filepath)
                print(n_columns, result, '\n')

            # If file doesn't exists, create it and append header
            append_header = not os.path.exists(PHD_OUTPUT_FILEPATH)

            # Save result into output file
            with open(PHD_OUTPUT_FILEPATH, 'a') as f:
                if append_header:
                    f.write(PROMPT_PHD+'\n')
                # Else, only append result
                f.write(result+'\n')
                f.close()
    #         count += 1
    #     if count == max_count:
    #         break
    # if count == max_count:
    #     break
```
